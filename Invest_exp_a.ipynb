```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import datetime
%matplotlib inline

# read Adj Close
adj_close = pd.read_csv("Stock_AdjClose.csv", index_col=0)
adj_close.index = adj_close.index.to_datetime()

# read data, Stock_NO d-type 맞춤, transpose
fin_df = pd.read_csv("year_eps.csv", index_col="Stock_NO")
fin_df.index = fin_df.index.astype(str)
fin_df.drop("Company_Name", axis=1, inplace=True)
fin_df = fin_df.transpose()

# 현재 존재하지 않는 기업 제외
fin_df.drop(["37620","60910","68870","66270","139200"], axis=1, inplace=True)

# index datetime 작업
fin_df.index = pd.to_datetime(fin_df.index)
fin_df = fin_df.shift(3, freq="M").shift(7, freq="D")

# 4월 10일 기준 3영업일 평균 종가 dataframe 만들기
before_row = fin_df.index.tolist() #PER 기준일
invest_row = fin_df.shift(3, freq="D").index.tolist() # 투자 기준일
after_row = fin_df.shift(9, freq="MS").index.tolist() # 정리 기준일
column_list = fin_df.columns.tolist()

before_close = pd.DataFrame()
invest_close = pd.DataFrame()
after_close = pd.DataFrame()
for i in column_list:
    before_close = before_close.append(pd.Series([adj_close[:x].ix[-3:, i].mean() for x in before_row], name=i))
    invest_close = invest_close.append(pd.Series([adj_close[:x].ix[-3:, i].mean() for x in invest_row], name=i))
    after_close = after_close.append(pd.Series([adj_close[:x].ix[-3:, i].mean() for x in after_row], name=i))

# dataframe 출력
before_close = before_close.transpose()
invest_close = invest_close.transpose()
after_close = after_close.transpose()
before_close.rename(lambda x: invest_row[x], inplace=True) # 날짜는 4월 10일 기준으로 맞춤, 계산하기 편하도록
invest_close.rename(lambda x: invest_row[x], inplace=True)
after_close.rename(lambda x: invest_row[x], inplace=True)
before_close.to_csv("before_close.csv", na_rep="NaN")
invest_close.to_csv("invest_close.csv", na_rep="NaN")
after_close.to_csv("after_close.csv_9m", na_rep="NaN")

after_close
```
